```python
# Booking Management System

user = "Ahmed"
pas = "12345"

# Booking data (date, day, bookings for 5 slots)
new_car_id = [
    ["13-04-2025", "Sunday",    [4, 6, 7, 8, 1]],
    ["14-04-2025", "Monday",    [2, 1, 8, 6, 4]],
    ["15-04-2025", "Tuesday",   [3, 1, 6, 6, 4]],
    ["16-04-2025", "Wednesday", [1, 6, 9, 8, 3]],
    ["17-04-2025", "Thursday",  [4, 7, 7, 10, 1]],
]

services = {
    1: "إصدار رخصة سيارة أول مرة",
    2: "تجديد رخصة سيارة",
    3: "إصدار رخصة قيادة أول مرة",
    4: "تجديد رخصة قيادة",
    5: "إصدار بدل فاقد"
}

slots = ["9:00", "10:00", "11:00", "12:00", "1:00"]

# -------------------------------
# Login Function
# -------------------------------
def login():
    print("Please log in")
    username = input("Username: ")
    password = input("Password: ")
    if username == user and  password == pas :
        print(f"Welcome {username}")
        return True
    else:
        print("Incorrect username or password")
        return False

# -------------------------------
# Display Services
# -------------------------------
def select_service():
    print("\nPlease select a service:")
    for key, value in services.items():
        print(f"{key}. {value}")
    while True:
        try:
            service_id = int(input("Enter service number: "))
            if service_id in services:
                return service_id
            else:
                print("Invalid service number.")
        except ValueError:
            print("Please enter a number.")

# -------------------------------
# Select Day
# -------------------------------
def select_day():
    print("\nPlease select the day:")
    days = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday"]
    for i, day in enumerate(days, 1):
        print(f"{i}. {day}")
    while True:
        try:
            day_index = int(input("Enter day number: "))
            if 1 <= day_index <= 5:
                return day_index - 1
            else:
                print("Invalid day number.")
        except ValueError:
            print("Please enter a number.")

# -------------------------------
# Show Available Slots
# -------------------------------
def show_slots(day_index):
    date, day, bookings = new_car_id[day_index]
    available = []
    print(f"\nAvailable slots for {day} ({date}):")
    for i, count in enumerate(bookings):
        remaining = 10 - count
        if remaining > 0:
            print(f"{i + 1}. {slots[i]} - {remaining} remaining")
            available.append(i)
    if not available:
        print("There are no available slots.")
    return available

# -------------------------------
# Select Time Slot
# -------------------------------
def select_slot(available):
    while True:
        try:
            slot_index = int(input("Select a slot number: ")) - 1
            if slot_index in available:
                return slot_index
            else:
                print("Invalid or unavailable slot.")
        except ValueError:
            print("Please enter a number.")

# -------------------------------
# Collect and Confirm Booking
# -------------------------------
def confirm_booking(service_id, day_index, slot_index):
    full_name = input("Enter your full name: ")
    id_number = input("Enter your ID number: ")

    date, day, bookings = new_car_id[day_index]
    bookings[slot_index] += 1

    print("\n--- Booking Confirmation ---")
    print(f"Name: {full_name}")
    print(f"ID: {id_number}")
    print(f"Service: {services[service_id]}")
    print(f"Date: {date}")
    print(f"Day: {day}")
    print(f"Slot: {slots[slot_index]}")
    print("----------------------------")

# -------------------------------
# Main Program
# -------------------------------
def main():
    if not login():
        return
    service_id = select_service()
    day_index = select_day()
    available = show_slots(day_index)
    if not available:
        return
    slot_index = select_slot(available)
    confirm_booking(service_id, day_index, slot_index)

main()
```
